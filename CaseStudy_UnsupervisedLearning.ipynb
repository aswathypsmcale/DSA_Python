```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Mall_Customers.csv')
```

```python
data.columns
```

```python
data=data.drop(['CustomerID','Genre','Age'],axis=1)
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(data,method='ward'))
# ward-- minimixation algorithm  metric--eucledian distance
plt.title('Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Eucledian distance')
plt.show()
```

```python
data.shape
```

```python
# assume cluster=5 in blue line
from sklearn.cluster import AgglomerativeClustering
```

```python
abc=AgglomerativeClustering(n_clusters= 5,metric='euclidean',linkage='ward')
y_abc=abc.fit_predict(data)
```

```python
y_abc
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil_ahc=silhouette_score(data,y_abc)
```

```python
sil_ahc
```
