```python
# Preprocessing
# 1. handle missing value---imputational methods
# 2. Feature scaling
# 3. Encode categorical column
# Missing value imputation -mean,median, mode
```

```python
# Feature scaling:  Normalization(0,1):--minmax,  or standaridization(normalization)
```

```python
# Normalization: doest not follow gaussian distribution
# standardization : column follow normal/gaussion
```

```python
# Robust scaler: when we know outliers in the data, we can use robust scaling to scale our data
# it uses IQR
```

```python
# How to encode categorical column
# 1 Nominal catgory( no rank/order)
# 2 ordinal catgory( Rank/order)

# withou tree based ML --One hot encoding
# Tree based ML --Label encoding

cat_col=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']
for col in cat_col:
  print(f'Distribution in {col}is:')
  print(df[col].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
# minimise label encoder
le=LabelEncoder()
# apply label encoding to each categorical column
for col in cat_col:
  df[col]=le.fit_transform(df[col])
# check the transformed col
df.head()
# this type of encoding cannot be used in linear model,svm or neural networks
```

```python
# creating one hot encoded features for working with non tree based algorithm
df_nontree=pd.get_dummies(df,columns=cat_col,drop_first=False)
df_nontree.head()
```

MODEL preparation

```python
# 1--split the data into target and features
x_nontree=df_nontree.drop('HeartDisease',axis=1)
y_nontree=df_nontree['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
X_train_nontree,X_test_nontree,Y_train_nontree,Y_test_nontree=train_test_split(x_nontree,y_nontree,test_size=0.2,random_state=42)

```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train_nontree,Y_train_nontree)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,accuracy_score
y_pred=lr.predict(X_test_nontree)
print(confusion_matrix(Y_test_nontree,y_pred))
print(classification_report(Y_test_nontree,y_pred))
print(accuracy_score(Y_test_nontree,y_pred))
```

Naive BAyers Model

```python
# use the concept of conditional probability
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,accuracy_score
gnb=GaussianNB()
gnb.fit(X_train_nontree,Y_train_nontree)

# predict
y_pred=gnb.predict(X_test_nontree)

# metrics
print(confusion_matrix(Y_test_nontree,y_pred))
print(classification_report(Y_test_nontree,y_pred))
print(accuracy_score(Y_test_nontree,y_pred))
```

```python
# apply scaling
from sklearn.preprocessing import StandardScaler
num_col= ['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak']
scaler=StandardScaler()
df[num_col]=scaler.fit_transform(df[num_col])
df.head()
```

```python
# split into x and Y
x=df.drop('HeartDisease',axis=1)
y=df['HeartDisease']

# plit into train and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
# fit decision tree and random forest on this
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)

y_pred_dt=dt.predict(X_test)

print(confusion_matrix(y_test,y_pred_dt))
print(classification_report(y_test,y_pred_dt))
print(accuracy_score(y_test,y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train,y_train)

y_pred_rf=rf.predict(X_test)

print(confusion_matrix(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
print(accuracy_score(y_test,y_pred_rf))
```

```python
# What happens next?
# get feature importance scores
importances = rf.feature_importances_

# sort the feature importance in descending order
indices=np.argsort(importances)[::1]
features = x.columns

# plot the importance feature
plt.figure(figsize=(10,6))
plt.bar(range(x.shape[1]),importances[indices],align='center')
plt.xticks(range(x.shape[1]),features[indices],rotation=90)
plt.show()
```
