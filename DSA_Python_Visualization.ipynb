```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplot()
```

```python
data=pd.read_csv('/content/medals_by_country_2016.csv')
data.head()
```

```python
data.shape
```

```python
plt.bar(data['Country'],data['Gold'])
plt.xticks(rotation=90)
plt.title('Gold Medals',color='green')
plt.xlabel('Country')
plt.ylabel('Gold Medals')
plt.show()
```

```python
plt.barh(data['Country'],data['Gold'],color='green')
#plt.xticks(rotation=90)
plt.title('Gold Medals',color='green')
plt.xlabel('Gold Medals')
plt.ylabel('Country')
plt.show()
```

```python
#Box plot
plt.boxplot(data['Gold'])
plt.show()
```

```python
plt.boxplot(data['Silver'])
plt.show()
```

```python
data1=data.set_index('Country')
plt.boxplot(data1)
plt.show()
```

```python
# Scatter plot
x=[2,4,5,7,9]
y=[1,5,6,8,10]
plt.scatter(x,y,color='purple')
x=[1,3,5,7,9]
y=[2,4,7,8,11]
plt.scatter(x,y,color='green')
plt.show()
```

```python
y=[30,40,50,100,60]
plt.pie(y)
plt.show()
```

```python
y=[30,40,50,100,60]
my_labels=['China','UK','Germany','India','USA']
#my_colors=['red','green','blue','yellow','orange']
my_colors=['b','m','c','g','r']
plt.pie(y,labels=my_labels,colors=my_colors)
#plt.legend(title='Countries')
plt.legend()
plt.title('Countries')
plt.show()
```

```python
# Histogram
plt.hist(data['Gold'])
plt.show()
```

```python
# Seaborn bargraph
sns.barplot(x=data['Country'],y=data['Gold'])
plt.xticks(rotation=90)
plt.show()
```
