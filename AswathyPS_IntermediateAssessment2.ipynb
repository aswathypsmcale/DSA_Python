```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
# Define numeric and categorical columns
numerical_cols = ['age', 'length_of_service', 'no_of_trainings', 'previous_year_rating', 'avg_training_score']
categorical_cols = ['department', 'region', 'education', 'gender', 'recruitment_channel']
```

```python
# Handle missing values and apply scaling/encoding
# Numeric features: Impute missing values with the median, then scale.
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
# Categorical features: Impute missing values with the mode, then encode using OneHotEncoder.
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Combine into one transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
# Apply preprocessing to the train and test sets separately
X_train = train_data.drop(columns='is_promoted')
y_train = train_data['is_promoted']
X_test = test_data
```

```python
# Apply preprocessing
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
# Modeling
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Train a Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_processed, y_train)
```

```python
# Train a Logistic Regression model
lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train_processed, y_train)
```

```python
# Evaluate the models
def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    print(f"Accuracy: {accuracy_score(y, y_pred):.4f}")
    print(classification_report(y, y_pred))

print("Random Forest Evaluation:")
evaluate_model(rf_model, X_train_processed, y_train)

print("Logistic Regression Evaluation:")
evaluate_model(lr_model, X_train_processed, y_train)
```

```python
# Fine-Tuning
from sklearn.model_selection import GridSearchCV

# Grid search for Random Forest hyperparameters
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42), param_grid=param_grid, cv=3, n_jobs=-1)
grid_search.fit(X_train_processed, y_train)

# Best model after grid search
best_rf_model = grid_search.best_estimator_
print(f"Best Random Forest Model: {grid_search.best_params_}")

# Evaluate the best model
evaluate_model(best_rf_model, X_train_processed, y_train)
```

```python
# Make predictions using the best model (let's assume it's Random Forest)
final_predictions = best_rf_model.predict(X_test_processed)

# Create the submission file
submission = pd.DataFrame({
    'employee_id': test_data['employee_id'],
    'is_promoted': final_predictions
})

# Save to CSV
submission.to_csv('final_submission.csv', index=False)
```
