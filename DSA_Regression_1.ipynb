```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('/content/tip-amount.xlsx')
```

```python

data.head(10)
```

```python
# find out correlation matrix
corre_matrix=data.corr()
```

```python
# plot data use heatmap
plt.subplots(figsize=(10,4))
sns.heatmap(corre_matrix,linewidth=0.2,annot=True)
```

```python
y=data['Observed tip amount(yi)']
x=data.drop('Observed tip amount(yi)',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
#random_state=42 universily approved
#test size we take ratio 25 and 75--train
```

```python
# train my model----training process called fit
from sklearn import linear_model
lr=linear_model.LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
#fit function used for training the model
```

```python
x_train
```

```python
y_train
```

```python
# test the model--predict function
pred=model.predict(x_test)
```

```python
pred
```

```python
x_test
```

```python
# original data
y_test
```

```python
# mean squared error valuee
from sklearn.metrics import mean_squared_error
print('MES is: \n',mean_squared_error(y_test,pred))
```

```python
from sklearn.metrics import r2_score
print('R2 score is: \n',r2_score(y_test,pred))
```
